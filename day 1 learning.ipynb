inheritance

```python
class Family:
    def __init__(self, father, mother):
        self.father = father
        self.mother = mother
    def relation(self):
          print("Parents:", self.father, self.mother)
class Big(Family):
    def __init__(self, father, mother, grandfather, grandmother):
        self.grandfather = grandfather
        self.grandmother = grandmother
        super().__init__(father, mother)
    def details(self):
        print("Parents:", self.father, self.mother)
        print("Grandparents:", self.grandfather, self.grandmother)
my_family = Big("x", "y", "a", "b")
my_family.relation()
my_family.details()
```

polymorphism

```python
class family:
  def intro(self):
    print("Im suwe")
  def study(self):
    print("mca")
class bro(family):
  def study(self):
    print("12th")
class cousin(family):
  def study (self):
    print("working")
obj1=family()
obj2=bro()
obj3=cousin()
obj1.intro()
obj1.study()
obj1.intro()
obj2.study()
obj1.intro()
obj3.study()
```

data abstraction

```python
from abc import ABC
class study(ABC):
  def score(self):
    pass
class internship(study):
  def score(self):
    print("good")
class extra(study):
  def score(self):
    print("avg")
obj1=study()
obj1.score()
obj2=internship()
obj2.score()
obj3=extra()
obj3.score()
```
